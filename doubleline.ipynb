```python
data=pd.read_csv('000001.csv',index_col='date')
data.head(10)
```

```python
ma_list=list()
close_series=data['close']
for i in range(5,len(close_series)):
    ma=close_series[i-5:i].mean()
    ma_list.append(ma)
df_ma=pd.DataFrame(ma_list,columns=['ma'])
df_ma.head(5)
```

```python
value=data['close'].rolling(5).mean()
value.head(10)
print(value.shift(-4).head(10))
```

```python
data['close'].rolling(5).max()
```

```python
df_ma=pd.DataFrame()
df_ma['ma5']=data['close'].rolling(5).mean()
df_ma['ma20']=data['close'].rolling(20).mean()
df_ma.head(50)
```

```python
df_ma.plot(figsize=(20,7))
```

```python
for i in range(21,df_ma.shape[0]):
    ma_fast_yesterday=df_ma['ma5'][i-1]
    ma_slow_yesterday=df_ma['ma20'][i-1]
    ma_fast_today=df_ma['ma5'][i]
    ma_slow_today=df_ma['ma20'][i]
    if(ma_fast_yesterday<ma_slow_yesterday and ma_fast_today>ma_slow_yesterday):
        print(data.index[i],'')
```

```python
df_new=pd.DataFrame()
df_new['ma_fast_today']=df_ma['ma5']
df_new['ma_fast_yesteray']=df_ma['ma5'].shift(1)
df_new['ma_slow_today']=df_ma['ma20']
df_new['ma_slow_yesterday']=df_ma['ma20'].shift(1)
df_new['value_today']=df_new['ma_fast_today']-df_new['ma_slow_today']
df_new['value_yesterday']=df_new['ma_fast_yesteray']-df_new['ma_slow_yesterday']
df_new['cross']=np.NaN
_index=df_new[(df_new['value_today']>0)&(df_new['value_yesterday']<0)].index
df_new.loc[_index,'cross']='gold_cross'
_index=df_new[(df_new['value_today']<0)&(df_new['value_yesterday']>0)].index
df_new.loc[_index,'cross']='death_cross'
df_new.head(50)



```
